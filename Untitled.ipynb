```python
# 텐서 형태로 변환하기 위한 TensorData 를 사용하기 위해 임포트 하였습니다.
# 텐서 형태로 변환한 이유는 병령 연산 및 학습 및 추론 속도를 향상시키기 위해 사용하였습니다.
import torch
from torch.utils.data import Dataset
```

```python
#데이터를 텐서(다차원 배열) 형태로 변환해주는 클래스 
class TensorData(Dataset):
  def __init__(self, x_data, y_data): # X_data 와 y_data 를 받아 각각 Float Tensor로 변환한뒤 저장
    self.x_data = torch.FloatTensor(x_data)# Double Tensor는 Float64로 변환 Float Tensor는 Float32로 변환 
    self.y_data = torch.FloatTensor(y_data) 
    self.len = self.y_data.shape[0]

  def __getitem__(self, index) : # 인덱스를 받아 해당 인덱스의 입력 및 출력 데이터 반환
    return self.x_data[index], self.y_data[index]

  def __len__(self): # 전체 데이터 길이 반환
    return self.len
```

```python
from sklearn.model_selection import train_test_split
```

```python
#scikit-learn 을 이용해서 학습 데이터(train)와 평가 데이터(test)로 나눈다. 
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.5) # hyper parameter 테스트 데이터 비율 사이즈

# train(학습) 데이터와 validation(검증) 데이터로 나눈다. 
X2_train, X_val, Y2_train, Y_val = train_test_split(X_train, Y_train) 

#학습 데이터랑 시험데이터를 배치 형태로 구축한다.
print(X2_train.shape)
print(X2_train.ndim)
print(X_val.shape)
print(Y2_train.shape)
print(Y2_train.ndim)
print(Y_val.shape)
trainsets = TensorData(X2_train, Y2_train) # 학습데이터를 생성. x_data(입력) 와 y_data(출력) 을 받아 floatTensor 로변환.
trainLoader = torch.utils.data.DataLoader(trainsets, batch_size = 32, shuffle=True) # 하이퍼 파라미터 배치 사이즈를 32(처리) 및 suffle로 데이터를 무작위로 섞는작업.

validationsets = TensorData(X_val, Y_val) #validation set을 위한 TensorData 생성.
validationLoader = torch.utils.data.DataLoader(validationsets, batch_size=32, shuffle=True) # DataLoader를 사용하여 검증 데이터셋을 배치 형태로 구축.

testsets = TensorData(X_test, Y_test) # 테스트 데이터들을 나눠준다. 
testloader = torch.utils.data.DataLoader(testsets, batch_size = 32, shuffle=True) # 테스트 데이터셋 비채 형태로 구축.
```

```python
# nn.Module 을 사용하기 위해 임포트 해줬습니다.
import torch.nn as nn
```

```python
# 모델MLP(multi Layer Perceptron) 다층 신경망 모델을 선택하였습니다.
# MPL : 압력층과 중간에 숨겨진 은닉충 그리고 결과를 출력하는 출력층으로 구분됩니다. 은닉층의 개수를 2개로 지정하여 진행했습니다.
```

```python
class Regressor(nn.Module):
    def __init__(self):
        super().__init__() # 모델 연산 정의 (부모 클래스 생성자 호출)
        self.fc1 = nn.Linear(4, 50, bias=True) # 입력층(11) -> 은닉층1(50)으로 가는 연산 bias=편향
        self.fc2 = nn.Linear(50, 30, bias=True) # 은닉층1(50) -> 은닉층2(30)으로 가는 연산
        self.fc3 = nn.Linear(30, 1, bias=True) # 은닉층2(30) -> 출력층(1)으로 가는 연산
        self.dropout = nn.Dropout(0.2) # 연산이 될 때마다 20%의 비율로 랜덤하게 노드를 없앤다(비활성화). (과적합[overfitting] 방지작업)

    def forward(self, x): # 모델 연산의 순서를 정의 ( 각 층의 연산과 활성화 함수를 순차적으로 적용한뒤 출력 반환 )
        x = F.relu(self.fc1(x)) # Linear 계산 후 활성화 함수 ReLU를 적용한다. (0보다 작을경우 0, 0보다 크면 그 값 그대로 반환)
        x = self.dropout(F.relu(self.fc2(x))) # 은닉층2에서 드랍아웃을 적용한다.(즉, 30개의 20%인 6개의 노드가 계산에서 제외된다.)
        x = F.relu(self.fc3(x)) # Linear 계산 후 활성화 함수 ReLU를 적용한다.  
      
        return x

  #drop out 은 과적합(overfitting)을 방지하기 위해 노드의 이부를 배제하고 계산하는 방식이기 때문에 출력층에 사용하면 안된다.
```

```python
# 위 MPL 모델에서 처음에 들어오면 fc1에서 입력을 받습니다. 변수가 4개이므로 4개에서 50개의 결과를 내보냅니다.
# 그 후 Fc2에서 50개를 다시 30개로 줄이는 과정을 Linear 함수를 이용해 진행합니다.
# 그 후 과적합을 방지하기 위해서 Dropout 함수를 이용해 한 epoch를 돌때마다 20% 비율로 랜덤하게 노드를 없애는 과정을 진행했습니다.
# Linear 함수를 통과한 후 활성하 함수로 Relu를 이용하였습니다. 
# Relu 함수를 이용한 이유는 음수의 영역에서는 0이다가 양수의 영역에서는 linear하게 증가하는 nonLinear함수로써,
# non Linear 함수를 곱해주지 않으면 결국 Linear해지는 문제도 존재하며 집 값이라는 데이터 특성 상 음수가 존재하지 않기때문에 적합하다고 생각되어 사용하였습니다.
```

```python
# 한 번 학습을 시킬때 Loss(정답과의 차이)를 구한 뒤 그 Loss를 줄이는 방향으로 학습을 시켰습니다.
# Loss를 구해줄 Loss함수 후 Loss를 줄이는 과정인 최적화 방법을 선택하였습니다.
# MSELoss로 Loss를 구했습니다. (정답값을 제곱한 Loss)

#최적화 방법으로는 Adam을 사용했으며 learning rate로는 0.00.1을 넣어주었습니다.
```

```python
# 최적화 알고리즘을 활용하기 위해 임포트하였습니다.
# 이때 Adam을 사용하였습니다.
import torch.optim as optim
```

```python
model = Regressor()
criterion = nn.MSELoss() # hyper parameter 활성화 함수 종류 평균제곱오차(MSE)로써 손실함수로 사용.(실제값 사이의 차이를 계산하여 학습하기 위해)

#lr == learning rate
#weight_decay는 L2 정규화에서의 penalty 정도

optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-7)#하이퍼 파라미터 lr weight decay optimizer 종류
# lr(learning rate) = 학습률 , weight_decay = 가중치 감소 [10e-7은 10의 음의 7승이며 모델의 복잡도를 제어하고 과적합을 방지하기위해 사용]
```

```python
# 학습마다 발생하는 Loss를 알기위해서 우선 Loss를 저장해줄 빈 배열을 선언하였습니다.
```

```python
# MSELoss를 활용하기 위해 임포트해줬습니다.
import torch.nn.functional as F
```

```python
loss_ = [] # 그래프를 그리기 위한 loss 저장용 리스트 
n = len(trainLoader) # 학습 데이터셋의 배치 수를 계산하여 변수 b에 저장.

for epoch in range(400): # 400번 학습을 진행한다.

    running_loss = 0.0 # 현재까지의 손실을 나타내는 변수 초기화.
    
    for i, data in enumerate(trainLoader, 0): # 무작위로 섞인 32개 데이터가 있는 배치가 하나 씩 들어온다.

        inputs, values = data # data에는 X, Y가 들어있다.

        optimizer.zero_grad() # 최적화 초기화(optimizer의 변화도를 초기화)
        
        outputs = model(inputs) # 모델에 입력값 대입 후 예측값 계산
        loss = criterion(outputs, values) # 예측값과 실제값을 비교하여 손실 함수 계산
        loss.backward() # 손실 함수 기준으로 역전파 설정 (손실에 대한 역전파 수행하여 모델 내 파라미터 변화도 계산)
        optimizer.step() # 역전파를 진행하고 가중치 업데이트( 계산된 변화도를 사용하여 모델의 파라미터 업데이트)
        running_loss += loss.item() # epoch 마다 평균 loss를 계산하기 위해 배치 loss를 더한다. 
 
    print(running_loss)
    loss_.append(running_loss/n) # MSE(Mean Squared Error) 계산[현재 epoch에서의 평균손실을 계산하여 손실리스트에 추가함]

        
print('Finished Training')
```

```python
# 모델 평가하기
#predictions = torch.tensor([]) # 예측 값을 저장하는 tensor
#actual = torch.tensor([]) # 실제 정답을 저장하는 tensor
```

```python
# RMSE를 계산하기 위해 임포트 하였습니다.
from sklearn.metrics import mean_squared_error
```

```python
def evaluation(dataLoader): # evaluation : 모델의 평가를 위한 함수 
    predictions = torch.tensor([]) # 예측 값을 저장하기 위한 (빈)tensor 입니다.
    actual = torch.tensor([]) # 실제값을 저장하기 위한 tensor

    with torch.no_grad(): # 모델이 평가 모드인 동안엔 그래디언트(함수의 기울기 또는 변화율)를 계산하지않도록 설정.
        model.eval() # 평가를 할때는 .eval()을 반드시 사용해야합니다.
        for data in dataLoader: # 데이터로더에서 데이터를 순회하면서 반복.
            inputs, values = data # 데이터의 입력값과 실제값 추출.
            outputs = model(inputs) # 모델에 입력값을 전달하여 예측값 계산.
            print(outputs)
            predictions = torch.cat((predictions, outputs), 0) # cat을 통해 예측값을 누적합니다.
            actual = torch.cat((actual, values), 0)
        predictions = predictions.numpy() # 텐서를 넘파이 배열로 변환 
        actual - actual.numpy() 
        #print(predictions)
        #print(actual)
        rmse = np.sqrt(mean_squared_error(predictions, actual)) # sklearn을 이용해서 RMSE를 합니다. (넘파이를 사용하여 예측과 실제값의 RMSE 계산)
        return rmse
    evaluation(trainLoader) # 위에서 정의한 함수를 호출하여 trainLoader를 전달하여 모델 평가.
        
```

```python
#평가 시 , eval()
#평가 시에는 온전한 모델로 평가를 해야하는데, .eval()이 아닌, .train()의 경우 드랍아웃이 활성화 되어있습니다.
#따라서 드랍아웃이나 배치 정규화 등과 같이 학습 시에만 사용하는 기술들을 평가 시에는 비활성화 해야합니다.
```

```python
train_rmse = evaluation(trainLoader) #학습 데이터의 RMSE ( 모델이 학습데이터에 대해 얼마나 정확히 예측하는지 )
vali_rmse = evaluation(validationLoader) #validation의 RMSE ( 모델이 검증더에티어 대해 얼마나 정확히 예측하는지 )
test_rmse = evaluation(testloader) # 모델이 테스트 데이터에 대해 얼마나 정확히 예측하는지 
print("Train RMSE: ", train_rmse)
print("Vali RMSE: ", vali_rmse)
print("Test RMSE: ", test_rmse)
```
